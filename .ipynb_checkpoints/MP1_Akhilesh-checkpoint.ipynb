# TASK 1] PROB 3)(b): Avg reaction time differentiated by location: 
## Highway - 1.48000 secs
## urban-street - 0.92865 secs

```python
t1=1.09
t2=0.9297703252032526
#H0: Mean time is same t1=t2
#Assume H0 is true and calculate P(t_mean=t2) (i.e.the p-value)

statistics.variance(dseng_df['ReactionTime'].dropna())
#using sample mean as population mean because the sample size is large (492)
sigma=math.sqrt(statistics.variance(dseng_df['ReactionTime'].dropna()))/math.sqrt(dseng_df['ReactionTime'].dropna().count())

#Calculating z-statistic
z=((t2-t1)/sigma)
print("Z value is " + str(z))
p_value = sp.ndtr(z)
print("P-value is " + str(p_value))
```

# Part (b) solution below:
## X = disengagement / mile (=1 if occured; = 0 if did not occur)
## W = weather (clr if clear, cldy = cloudy)


$$ P (X = 1 | W=cldy) = \frac {P (W=cldy | X=1) * P (X=1)}{P(W=cldy)} $$

```python
z=dseng_df.groupby('Weather')['Month'].count()

P1=z[1]/(z[0]+z[1]) # P (W=cldy | X=1)
print("Probablity it is cloudy, given a disengagement has occured " + str(P1))

miles=summary_month_df['miles driven in autonomous mode'].sum()
print("Total number of miles driven are " + str(miles))
n_dseng=1024
print("Total number of disengagements occured are " + str(n_dseng))

P2=n_dseng/miles # P (X=1) 

P3=1-0.72 # P(W=cldy)

answer = P1*P2/P3
round(answer,5)
```

# Part (c) solution below:

$$ P (X = 1 | W=clr) = \frac {P (W=clr | X=1) * P (X=1)}{P(W=clr)} $$

```python
z=dseng_df.groupby('Weather')['Month'].count()

P1=z[0]/(z[0]+z[1]) # P (W=clr | X=1)
print("Probablity it is clear, given a disengagement has occured " + str(P1))

miles=summary_month_df['miles driven in autonomous mode'].sum()
print("Total number of miles driven are " + str(miles))
n_dseng=1024
print("Total number of disengagements occured are " + str(n_dseng))

P2=n_dseng/miles # P (X=1) 

P3=0.72 # P(W=clr)

answer = P1*P2/P3
round(answer,4)
```

# Part (d) solution below:
## X = automatic disengagement / mile (= 1 if occured; = 0 if did not occur)
## W = weather (clr if clear, cldy = cloudy)


$$ P (X = 1 | W=cldy) = \frac {P (W=cldy | X=1) * P (X=1)}{P(W=cldy)} $$

```python
z=dseng_df.groupby(['TypeOfTrigger','Weather']).count()

P1=z.iloc[1,0]/(z.iloc[0,0]+z.iloc[1,0]) # P (W=cldy | X=1)
print("Probablity it is cloudy, given an automatic disengagement has occured " + str(P1))

miles=summary_month_df['miles driven in autonomous mode'].sum()
print("Total number of miles driven are " + str(miles))

n_auto_dseng=dseng_df.groupby('TypeOfTrigger').count().loc['automatic','Month']
print("Total number of automatic disengagements occured are " + str(n_auto_dseng))

P2=n_auto_dseng/miles # P (X=1) 

P3=1-0.72 # P(W=cldy)

answer = P1*P2/P3
round(answer,5)
```

```python
z=dseng_df.groupby(['TypeOfTrigger','Weather']).count()

P1=z.iloc[0,0]/(z.iloc[0,0]+z.iloc[1,0]) # P (W=clr | X=1)
print("Probablity it is clear, given an automatic disengagement has occured " + str(P1))

miles=summary_month_df['miles driven in autonomous mode'].sum()
print("Total number of miles driven are " + str(miles))

n_auto_dseng=dseng_df.groupby('TypeOfTrigger').count().loc['automatic','Month']
print("Total number of automatic disengagements occured are " + str(n_auto_dseng))

P2=n_auto_dseng/miles # P (X=1) 

P3=0.72 # P(W=clr)

answer = P1*P2/P3
round(answer,5)
```
